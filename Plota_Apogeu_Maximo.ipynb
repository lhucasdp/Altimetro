```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# É necessário subir o arquivo '.txt' ou passar o path do local
# OBS: o skiprows=1 para ignorar a primeira linha que é um divisor (=================)
df = pd.read_csv("dados_altitude.txt", sep="\t", skiprows=1)  # colunas: tempo_ms, altitude
```

```python
linha_apogeu = df.loc[df["altitude_m"].idxmax()] # Encontra o apogeu
```

```python
print("Apogeu encontrado:")
print(f"Altura máxima: {linha_apogeu['altitude_m']:.2f} metros")
print(f"Tempo: {linha_apogeu['tempo_ms']} ms")
```

```python
# Alteração e plotagem do gráfico
plt.figure(figsize=(10,5))
plt.plot(df["tempo_ms"]/1000, df["altitude_m"], color="blue", label="Altitude")
plt.title("Altitude vs Tempo")
plt.xlabel("Tempo (s)")
plt.ylabel("Altitude (m)")
plt.legend()
plt.grid(True)
plt.scatter(linha_apogeu["tempo_ms"]/1000, linha_apogeu["altitude_m"], color="red", label="Apogeu")
```
